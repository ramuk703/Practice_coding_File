Box Plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt. show() 
```

```python
plt.boxplot(x,vert=False)
plt.show()
```

```python
plt.boxplot(x,widths=0.8)
plt.show()
```

```python
plt.boxplot(x,labels=["Python"],patch_artist=True)
plt.show()
```

```python
plt.boxplot(x,labels=["Python"],patch_artist=True,showmeans=True)
plt.show()
```

```python
y = [80,90,10,20,30,50,50,200]
plt.boxplot(y)
plt.show()
```

```python
y = [80,90,10,20,30,50,50,200]
plt.boxplot(y,whis=3)
plt.show()
```

```python
y = [80,90,10,20,30,50,50,200]
plt.boxplot(y,sym="g+")
plt.show()
```

```python
y = [80,90,10,20,30,50,50,200]
x = [10,20,30,40,50,60,70,89]
s = [10,57,68,90,43,64,45,100]
z = [x,y,s]
plt.boxplot(z,labels=["Python","DSA","c++"],boxprops=dict(color="r"),capprops=dict(color="r"))

plt.show()
```
