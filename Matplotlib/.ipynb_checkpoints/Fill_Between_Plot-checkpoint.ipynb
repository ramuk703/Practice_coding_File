Fill Between Plot in Matplotlib Python

```python
import matplotlib.pyplot as plt
x = [1,2,3,5,6,7,8]
y = [1,2,3,5,6,7,8]
plt.plot(x,y,color="Orange")
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.title("Python Programing",fontsize=15)
plt.fill_between(x=[3,6],y1=3,y2=6)
plt.show()
```

```python
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="y")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
z = np.array([1,2,3,4,5,6])
a = np.array([1,2,3,4,5,6])
plt.plot(z,a,color="r")
plt.fill_between(z,a,where=(z>=2) & (z<=4))
plt.show()
```
