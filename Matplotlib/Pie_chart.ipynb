How to Create a Pie Chart in Python ?

```python
import matplotlib.pyplot as plt
x = ["India","Australia","London","Germany","Brazil","Spain"]
y =[90,80,70,60,50,40]
plt.pie(y,labels=x,autopct="%0.1f%%")
plt.legend()
plt.show()
```

```python
# Pie plot with percentage arrugmnets
import matplotlib.pyplot as plt
x = ["India","Australia","London","Germany","Brazil","Spain"]
y =[90,80,70,60,50,40]
ex=[0.0,0.0,0.0,0.5,0.0,0.0]
plt.pie(y,labels=x,explode=ex,autopct="%0.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["India","Australia","London","Germany","Brazil","Spain"]
y =[90,80,70,60,50,40]
ex=[0.0,0.2,0.0,0.4,0.0,0.0]
plt.pie(y,labels=x,explode=ex,autopct="%0.5f%%")
plt.show()
```

```python
# Pie plot appy shadown and radius arrangement

import matplotlib.pyplot as plt
x = ["India","Australia","London","Germany","Brazil","Spain"]
y =[90,80,70,60,50,40]
ex=[0.0,0.0,0.0,0.4,0.0,0.0]
plt.pie(y,labels=x,explode=ex,autopct="%0.1f%%",shadow=True,radius=2)
plt.legend(loc=1)
plt.show()
```

```python
#  Pie plot font size increase or decrease with pie chart direction
import matplotlib.pyplot as plt
x = ["India","Australia","London","Germany","Brazil","Spain"]
y =[90,80,70,60,50,40]
ex=[0.0,0.0,0.0,0.3,0.0,0.0]
plt.pie(y,labels=x,explode=ex,autopct="%0.1f%%",startangle=90,labeldistance=1.1,textprops={"fontsize":13})
plt.show()
```

```python
#  Pie plot clockwise rotate
import matplotlib.pyplot as plt
x = ["India","Australia","London","Germany","Brazil","Spain"]
y =[90,80,70,60,50,40]
ex=[0.0,0.0,0.0,0.3,0.0,0.0]
plt.pie(y,labels=x,explode=ex,autopct="%0.1f%%",startangle=90,labeldistance=1.1,counterclock=False,shadow=True,wedgeprops={"linewidth":10,"edgecolor":"yellow"}
       ,rotatelabels=True)

plt.show()
```

```python
#  Pie plot font size increase or decrease with pie chart direction
import matplotlib.pyplot as plt
x = ["India","Australia","London","Germany","Brazil","Spain"]
y =[90,80,70,60,50,40]
z =[50,30,50,40,65,70]

plt.pie(y,labels=x,startangle=90,labeldistance=1.1,radius=1.5)
plt.pie([1],colors="w")

plt.show()
```
