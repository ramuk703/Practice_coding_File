Area vs stack plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
area = [20,10,20,30,45,65,40]
area1 = [23,34,44,53,23,43,34]
area2 = [53,34,34,23,43,23,43]
plt.stackplot(x,area,area1,area2)
plt.show()
```

```python
l = ["area1","area2","area3"]
plt.stackplot(x,area,area1,area2,labels=l)
plt.legend()
plt.show()
```

```python
l = ["area1","area2","area3"]
plt.xlabel("X-axis",fontsize=15)
plt.ylabel("Y-axis",fontsize=15)
plt.title("Programing",fontsize=15)
plt.stackplot(x,area,area1,area2,labels=l,colors=["r","b","Orange"])
plt.legend()
plt.show()
```
