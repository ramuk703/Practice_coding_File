Multiple Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["MTH","Science","Hindi","Social Science","English"]
y =[40,60,40,30,70]
z =[80,20,30,60,40]
plt.xlabel("Name of Subjects",fontsize=15)
plt.ylabel("No.of Marks",fontsize=15)
plt.title("Students Reports",fontsize=15)
width= 0.4
p = np.arange(len(x))
p1 =[j+width for j in p]
plt.bar(p,y,width,color="r",label="Proerty1")
plt.bar(p1,z,width,color="y",label="Property2")
plt.legend()

plt.xticks(p+width/2,x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["MTH","Science","Hindi","Social Science","English"]
y =[40,60,40,30,70]
z =[80,20,30,60,40]
plt.xlabel("Name of Subjects",fontsize=15)
plt.ylabel("No.of Marks",fontsize=15)
plt.title("Students Reports",fontsize=15)
width= 0.4
p = np.arange(len(x))
p1 =[j+width for j in p]
plt.bar(p,y,width,color="r",label="Proerty1")
plt.bar(p1,z,width,color="y",label="Property2")
plt.legend()

# subjects rotations property
plt.xticks(p+width/2,x,rotation=20)
plt.show()
```
