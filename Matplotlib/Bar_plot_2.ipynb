BAR PLOT 1

```python
import matplotlib.pyplot as plt
x = ["Math","Science","English","Hindi","Social Science",]
y = [80,90,100,54,64]
plt.bar(x,y)
plt.show()
```

```python
# Bar plot Name arrangement in out side of frame
plt.bar(x,y)
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
# Bar plot width size arrangement
plt.bar(x,y,width=0.5)
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
# Bar plot color arrangement
c=["r","b","g","c","y"]
plt.bar(x,y,width=0.5,color=c)
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
# Bar plot name arrangement align center
c=["r","b","g","c","y"]
plt.bar(x,y,width=0.5,color=c,align="center")
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
# Bar plot edge color/ border color
c=["r","b","g","c","y"]
plt.bar(x,y,width=0.5,color=c,edgecolor="r")
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
# Bar plot border width / line width
c=["r","b","g","c","y"]
plt.bar(x,y,width=0.5,color=c,edgecolor="m",linewidth=4)
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
# Bar plot border style arragnement
c=["r","b","g","c","y"]
plt.bar(x,y,width=0.5,color=c,edgecolor="m",linewidth=4,linestyle=":")
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
# Bar plot transparent color arrangement
c=["r","b","g","c","y"]
plt.bar(x,y,width=0.5,color=c,edgecolor="m",linewidth=4,linestyle=":",alpha=0.2)
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```

```python
c=["r","b","g","c","y"]
plt.bar(x,y,width=0.5,color=c,edgecolor="m",linewidth=4,alpha=0.2,label="Bar_Graph")
plt.legend()
plt.xlabel("Name of Subjects",fontsize=20)
plt.ylabel("Number of Marks",fontsize=15)
plt.title("Student Reports",fontsize=15)
plt.show()
```
