Stems Plots with Matplotlib

```python
import matplotlib.pyplot as plt
x = [1,2,3,5,6,7,8]
y = [3,2,1,3,2,5,6]
plt.stem(x,y)
plt.show()
```

```python
# Change point color by markerfmt parameter and line style change for linefmt dot through
x = [1,2,3,5,6,7,8]
y = [3,2,1,3,2,5,6]
plt.stem(x,y,linefmt="-.",markerfmt="r",bottom=2)
plt.show()
```

```python
# Change point color by markerfmt parameter and line style change for linefmt dot through
x = [1,2,3,5,6,7,8]
y = [3,2,1,3,2,5,6]
plt.stem(x,y,linefmt="-.",markerfmt="r",bottom=2,basefmt="orange",label="Python Tutorial")
plt.legend()
plt.show()
```

```python
# Change point color by markerfmt parameter and line style change for linefmt dot through
x = [1,2,3,5,6,7,8]
y = [3,2,1,3,2,5,6]
plt.stem(x,y,linefmt=":",markerfmt="r",bottom=2,basefmt="orange",label="Python Tutorial",orientation="horizontal")
plt.legend()
plt.show()
```
