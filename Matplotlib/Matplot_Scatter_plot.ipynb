Matplotlib Scatter Plot

```python
import matplotlib.pyplot as plt
x = ["Phase_1","Phase_2","Phase_3","Phase_4","Phase_5","Phase_6","Phase_7"]
y = [100,20,39,43,42,10,20,]
# y = [23,53,64,64,22,11,12,21,32,43,45,54]
# c =["r","y"]
plt.title("Scatter Plot",fontsize=12)
plt.xlabel("Day",fontsize=14)
plt.ylabel("No of Voting in Multiple Phases",fontsize=12)
color=["r","g","m","c","b","y","b"]
size=[500,200,60,20,10,50,90]
plt.scatter(x,y,c=color,s=size)
plt.show()
```

```python
# Scatter Change in Start form

plt.title("Scatter Plot",fontsize=12)
plt.xlabel("Day",fontsize=14)
plt.ylabel("No of Voting in Multiple Phases",fontsize=12)
color=["r","g","m","c","b","y","b"]
size=[500,200,60,20,10,50,90]

plt.scatter(x,y,c=color,s=size,marker="*")
plt.show()
```

```python

import matplotlib.pyplot as plt
day = [10,20,30,29,40,40,20]
no = [100,20,39,43,42,10,20,]
# y = [23,53,64,64,22,11,12,21,32,43,45,54]
# c =["r","y"]

colors=[50,60,40,60,80,90,10]
sizes=[500,200,60,20,10,50,90]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis",alpha=0.6)
plt.colorbar()
plt.title("Scatter Plot",fontsize=12)
plt.xlabel("Day",fontsize=14)
plt.ylabel("No",fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,8,6,7,3,6,9]
colors=[50,60,40,60,80,90,10]
sizes=[500,200,60,20,10,50,90]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
t=plt.colorbar()
t.set_label("Color Bar")
plt.title("Scatter Plot",fontsize=12)
plt.xlabel("day",fontsize=14)
plt.ylabel("no",fontsize=12)

plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,8,6,7,3,6,9]
no2 =[6,3,7,8,9,10,22]
colors=[50,60,40,60,80,90,10]
sizes=[500,200,60,20,10,50,90]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
plt.scatter(day,no2,c="r",s=sizes,cmap="viridis")
t=plt.colorbar()
t.set_label("Color Bar")
plt.title("Scatter Plot",fontsize=12)
plt.xlabel("day",fontsize=14)
plt.ylabel("no",fontsize=12)

plt.show()
```
