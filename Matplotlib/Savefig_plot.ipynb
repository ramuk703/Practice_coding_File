How to Use Savefig in Matplotlib?

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,90]
y = [20,10,50,20,40,80,90,10,60]
c=["r","b","g","m","Orange","c","y","r","b"]
plt.bar(x,y,color=c,width=5)
plt.savefig("Bar_Graph5",dpi=100,facecolor="b",transparent=True,bbox="inc")
plt.show()
```
