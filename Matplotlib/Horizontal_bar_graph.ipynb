HORIZONTAL BAR GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
x =["Math","Science","English","Social Science","Hindi"]
y =[50,30,60,70,80]
z =[80,30,20,60,90]
width=0.4
p = np.arange(len(x))
p1 =[j+width for j in p]
plt.xlabel("Subjects Name",fontsize=15)
plt.ylabel("No of Marks",fontsize=15)
plt.title("Students Reports",fontsize=15)
plt.barh(p,y,width,color="r",label="Property1")
plt.barh(p1,z,width,color="y",label="Property2")
plt.legend()
plt.xticks(p+width/2,x)
plt.show()
```
