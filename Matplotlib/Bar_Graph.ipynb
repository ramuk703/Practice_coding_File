How to Plot Bar Graph in Matplotlib ?

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Math","Science","Social Science","English","Hindi"]
y =[35,64,56,76,86]
z =[12,21,24,54,66]
r =[23,45,64,64,88]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
width1=0.4
l=np.arange(len(x))
r1=[k+width1 for k in l]

plt.xlabel("Subjects Name",fontsize=15)
plt.ylabel("No of Marks",fontsize=15)
plt.title("Students Reports",fontsize=15)

plt.bar(p,y,width,align="center",color="r",label="Property1")
plt.bar(p1,z,width,align="center",color="y",label="Property2")
plt.bar(r1,r,width,align="center",color="g",label="Property2")

plt.xticks(p+width/3,x,rotation=20)
plt.legend()
plt.show()
```
