```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]

plt.subplot(2,2,1)
plt.plot(x,y,color="r")


plt.pie([1],colors="r")
x1 = [10,20,30,40,50,60]

plt.pie(x)

z = [10,20,30,40,50,60]
m = [20,30,40,50,60,70]

plt.bar(z,m,width=5)
plt.show()
```
